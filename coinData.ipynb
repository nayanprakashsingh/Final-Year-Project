```python
import pandas as pd
```

```python
df = pd.DataFrame({'Name':list('aabbef'),
                   'A':[4,5,4,5,5,4],
                   'B':[7,8,9,4,2,3],
                   'C':[1,3,5,7,1,0]}, columns = ['Name','A','B','C'])
```

```python
print(df)
```

```python
d = dict(tuple(df.groupby('Name')))
```

```python
for k,v in d.items():
    v.to_csv(k+'.csv',index=False)
```
